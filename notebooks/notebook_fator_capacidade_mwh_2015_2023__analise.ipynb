###Optar por utilizar pd.NaN em vez de None ou pd.Na para preencher valores nulos é uma escolha vantajosa, uma vez que o pd.NaN é mais compatível com todas as bibliotecas Python, oferecendo uma integração mais suave em operações numéricas e estatísticas no pandas E Além disso, essa escolha se aplica não apenas a colunas numéricas, mas também a colunas do tipo string. Para colunas do tipo string, utilizar pd.NaN é mais favorável do que pd.Na ou None, uma vez que mantém a consistência e a capacidade de manipulação em todo o ecossistema Python.
###A mediana é uma medida de tendência central que é menos sensível a valores discrepantes do que a média, zero, pd.NaN ou pd.Na. Isso significa que a mediana é menos propensa a ser distorcida por valores outliers. Além disso, a mediana preserva a distribuição original dos valores, o que resulta em preenchimentos mais robustos e representativos.

Remover dados duplicados, corrigir erros de digitação, tratar dados inconsistentes etc.

````python
df_fator_cap_analise.head
````

````python
#Verifique se há algum valor duplicado
if df_fator_cap_analise.duplicated().any():
    print('Há valores duplicados.')
else:
  print('Não há valores duplicados.')
````

````python
#Visualizando as colunas que apresentam valores ausentes (nulos)
df_fator_cap_analise.isnull().sum()
````

````python
#Verificando os tipos de dados presentes no dataframe
print(df_fator_cap_analise.dtypes)
````

````python
#Preenchendo as colunas com valores ausentes utilizando pd.NaN porque é um valor especial reconhecidoo por todas as bibliotecas Python
# Loop através das colunas do DataFrame e preencher com pd.NAN. O pd.NaN é um valor especial que é reconhecido por todas as bibliotecas Python e
# e o tipo de valor especial mais adequado para preencher os valores ausentes em colunas do tipo String(object), em relação ao "pd.na" ou "None"
#para esse caso específico deste dataframe.
for coluna in df_fator_cap_analise.columns:
  if df_fator_cap_analise[coluna].dtype == "float64":
    df_fator_cap_analise[coluna].fillna(df_fator_cap_analise[coluna].median(), inplace=True)
  elif df_fator_cap_analise[coluna].dtype =="object":
    df_fator_cap_analise[coluna].replace( np.NaN, inplace=True)

````

````python
#Verificando se os valores ausentes nas colunas foram preenchidos
df_fator_cap_analise.isnull().sum()
````

### Filtros e agrupamentos
Criar novas variáveis com base nos dados originais para permitir uma análise mais detalhada e avançada.

````python
# Agrupamento por Ano e tipo de usina
grouped = df_fator_cap_em_analise.groupby(['Data', 'Nom_tipousina']).agg({
    'Val_geracaoprogramada': 'sum',
    'Val_geracaoverificada': 'sum',
    'Val_capacidadeinstalada': 'sum'
}).reset_index()

display (grouped)
````

````python
# Agrupando por região a capacidade instalada
grouped = df_fator_cap_em_analise.groupby('Nom_subsistema')['Val_capacidadeinstalada'].sum().reset_index()

display (grouped)
````

# Visualização

````python
# Gráfico de Pizza
plt.figure(figsize=(8, 8))
df_pie = df_fator_cap_em_analise.groupby('Nom_tipousina')['Val_capacidadeinstalada'].sum()
df_pie.plot(kind='pie', autopct='%.2f%%')
plt.title('Distribuição de Capacidade Instalada por Tipo de Usina')
plt.ylabel('')
plt.tight_layout()
plt.show()
````

````python
# Calcular a capacidade instalada total das fontes de geração
capacidade_instalada = df_fator_cap_em_analise.groupby('Nom_tipousina')['Val_capacidadeinstalada'].sum()

# Calcular a capacidade instalada total
total_capacidade_instalada = capacidade_instalada.sum()

# Calcular a porcentagem de cada fonte
porcentagem_capacidade_instalada = (capacidade_instalada / total_capacidade_instalada) * 100

# Criar o gráfico de barra horizontal
plt.figure(figsize=(10, 6))
porcentagem_capacidade_instalada.sort_values().plot(kind='barh')
plt.title('Distribuição das Fontes de Geração pela Capacidade Instalada (%)')
plt.xlabel('Porcentagem (%)')
plt.ylabel('Fonte de Geração')

# Adicionar os valores percentuais como rótulos nas barras
for index, value in enumerate(porcentagem_capacidade_instalada.sort_values()):
    plt.text(value, index, f'{value:.2f}%', va='center')

plt.tight_layout()
plt.show()





````
